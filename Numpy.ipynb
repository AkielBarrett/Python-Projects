# Creating Arrays


```python
a1 = np.array([3, 5, 7, 3]) # creates a numpy array
a2 = np.zeros (10) # creates an array of zeros
a3 = np.ones (10) # creates an array of ones
a4 = np.random.random(10) # random.random generates numbers from a uniform distribution 
a5 = np.random.randn(10) # random.randn generates numbers from a normal distribution
a6 = np.linspace(0, 10, 100) # creates an array of evenly spaced numbers over a specified interval eg. 0 to 10 of 0.02
a7 = np.arange(0, 10, 0.02) # creates an array of evenly spaced numbers with a step size
```

# Array Indexing/Slicing 

```python
a1 = np.array([2, 4, 6, 8, 10])
```

```python
a1[2] # returns the third element of the array
```

```python
a1[2:] # returns the third element to the end of the array
```

```python
a1[-2:] # returns the last two elements of the array
```

```python
a1[1:-2] # returns the second element to the second last element of the array
```

**Boolean indexing**

```python
a1>3 # returns a boolean array
```

```python
a1[a1>3] # returns the elements of the array that are greater than 3
```

## lambda and vectorize

```python
# lambda function to get the first letter of each name
f = lambda s: s[0] # returns the first letter of each name
```

```python
f('animal') # returns the first letter of the string
```

```python
# np.vectorize to apply the function to each element of the array
np.vectorize(lambda s: s[0])(names) # returns the first letter of each name in the array
```

```python
# np.vectorize(lambda s: s[0])(names) is an array
```

```python
np.vectorize(lambda s: s[0])(names)=='D' # returns a boolean array of the names that start with D
```

```python
first_letter_D = np.vectorize(lambda s: s[0])(names)=='D' # returns a boolean array of the names that start with D
```

```python
names[first_letter_D] # returns the names that start with D
```

```python
names = np.array(['John', 'Mary', 'Peter', 'Paul', 'Anna']) # creates an array of names
first_letter_p = np.vectorize(lambda s: s[0])(names)=='P' # returns a boolean array of the names that start with P
names[first_letter_p] # returns the names that start with P
```
